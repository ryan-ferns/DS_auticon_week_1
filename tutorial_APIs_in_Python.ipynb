```python
import requests as re
import os
```

```python
client_id = os.environ["FOURSQUARE_API_ID"]
client_secret = os.environ["FOURSQUARE_API_SECRET"]
location = "Montreal,Canada"
v = "20200731"
```

```python
url = "https://api.foursquare.com/v2/venues/search?near="+location+"&client_id="+client_id+"&client_secret="+client_secret+"&v="+v
```

```python
res = re.get(url)
```

```python
print(res)
```

```python
print(res.text)
```

```python
import requests as re
import os
```

```python
params_T = {'near': "Toronto,Canada", 'client_id': os.environ["FOURSQUARE_API_ID"], 
                "client_secret": os.environ["FOURSQUARE_API_SECRET"],
                'v': "20200731"}
```

```python
response = re.get(
    "https://api.foursquare.com/v2/venues/search",
    params = params_T,
)
```

```python
print(response)
```

```python
response.json()
```
